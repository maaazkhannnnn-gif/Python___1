```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import ast

file = pd.read_excel('C:/Data_/excel/data_jobs_salary_all.xlsx',sheet_name='maaz_1')
file['job_posted_month']=pd.to_datetime(file['job_posted_date']).dt.strftime('%b')
file['job_skills']=file['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)

print(file.info())
```

```python
file_pie=file.copy()
file_pie
```

```python
file_pie['job_work_from_home'].value_counts().plot(kind='pie',startangle=90,autopct='%1.1f%%')  #what is "%1.1f%%"
plt.title('Work from Home Status')
plt.ylabel("")
plt.tight_layout()
plt.show()

```

```python
file[['job_work_from_home','job_no_degree_mention','job_health_insurance']]
fig, ax=plt.subplots(1,3)

dict_col={'job_work_from_home':'Work_Frome_home',
          'job_no_degree_mention':'Job_Degree_Mention',
          'job_health_insurance':'Health_Insurance'}

for i,(column,title) in enumerate(dict_col.items()):
    ax[i].pie(file[column].value_counts(),startangle=90,autopct='%1.1f%%',labels=['False','True'])
    ax[i].set_title(title)

plt.tight_layout()

```

```python
file['job_health_insurance'].value_counts().plot(kind='pie',startangle=90,autopct='%1.1f%%',colormap='viridis')
plt.title('Health Insurance')
plt.tight_layout()
plt.show()
```
