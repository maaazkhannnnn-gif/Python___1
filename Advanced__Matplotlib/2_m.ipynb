```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import ast

file = pd.read_excel('C:/Data_/excel/data_jobs_salary_all.xlsx',sheet_name='maaz_1')
file['job_posted_month']=pd.to_datetime(file['job_posted_date']).dt.strftime('%b')


print(file.info())
```

```python
file
```

```python
import ast

file['job_skills']=file['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
file
```

```python
file_explode=file.explode('job_skills')
file_explode=file_explode.dropna(subset=['job_skills'])
file_explode['job_skills'].head(40)
```

```python
file_skill_counts=file_explode['job_skills'].value_counts()
file_skill_counts
```

```python

file_explode    #All datasets with explode(job_skills)

Skill_count=file_explode.groupby(['job_title_short','job_skills']).size()
type(Skill_count)
```

```python

new_count=Skill_count.reset_index(name='skills_count')
new_count=new_count.sort_values(by='skills_count',ascending=False)
new_count
```

```python
job_title='Data Analyst'
top_skills=10

new_count=new_count[new_count['job_title_short']==job_title].head(top_skills)
file_skill=new_count.plot(kind='barh',x='job_skills',y='skills_count')
plt.legend("")
plt.gca().invert_yaxis()
plt.show()

```
