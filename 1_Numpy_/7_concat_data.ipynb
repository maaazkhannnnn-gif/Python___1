```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

                   
file = pd.read_excel('C:/Data_/excel/data_jobs_salary_all.xlsx',sheet_name='maaz_1')
file['job_posted_month']=file['job_posted_date'].dt.strftime('%b')
file['month_no']=pd.to_datetime(file['job_posted_date'],format='%b').dt.month
print(file)
```

```python
months=file['job_posted_month'].unique()
dict_month={month:file[file['job_posted_month']==month] for month in months}
dict_month['Jan']
```

```python
quar_1=pd.concat([dict_month['Jan'],dict_month['Feb'],dict_month['Mar']],ignore_index=True)
a=quar_1['job_posted_month'].value_counts().plot(kind='bar')
a
```
