##Data Management

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

file = pd.read_excel('C:/Data_/excel/data_jobs_salary_all.xlsx',sheet_name='maaz_1')
print(file.sample(10,random_state=42))
```

```python
file_filtered=file.copy()
file_filtered
```

```python
chart=file.pivot_table(index='job_title_short',aggfunc='size').sort_values()
chart.plot(kind='barh')
plt.show()
```

```python
pivot_table=file.pivot_table(values='salary_year_avg',index='job_country',aggfunc='size',columns='job_title_short')
pivot_table
```

```python
top_countries=file['job_country'].value_counts().head(6).index
file_job_year_salary=file.pivot_table(values='salary_year_avg',index='job_country',aggfunc='median',columns='job_title_short')
file_job_year_salary=file_job_year_salary.loc[top_countries]
job_title=['Data Analyst','Data Engineer','Data Scientist']
#file_job_year_salary=file_job_year_salary[job_title]
file_job_year_salary[job_title].plot(kind="barh")
plt.legend(fontsize=6)
plt.xlabel('Meidan Salary($USD)')
plt.show()
```
