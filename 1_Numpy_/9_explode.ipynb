```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

file = pd.read_excel('C:/Data_/excel/data_jobs_salary_all.xlsx',sheet_name='maaz_1')
file['job_posted_month']=pd.to_datetime(file['job_posted_date'])
file['job_posted_month']=file['job_posted_date'].dt.strftime('%b')



print(file.info())
```

```python
import ast
file['job_skills'] = file['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
file_exploded=file.explode('job_skills')
count_skills=file_exploded['job_skills'].value_counts().head(10).plot(kind='barh')
count_skills
plt.gca().invert_yaxis
plt.show()


```

```python
Job_title='Data Analyst'
Top_Skills=20

file_exploded_final=file_exploded[file_exploded['job_title_short']==Job_title].head(Top_Skills)
file_exploded_final.value_counts('job_skills').plot(kind='bar')
```
